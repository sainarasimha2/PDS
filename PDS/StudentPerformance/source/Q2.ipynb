```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sd_data=pd.read_csv("/content/student.csv")
```

```python
# Visualization 1: Bar Chart - Gender Distribution
gender_counts = sd_data['gender'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='race/ethnicity', y='math score', data=sd_data)
plt.title("Math Scores by Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Math Score")
plt.show()
```

```python
# Visualization 3: Scatter Plot - Math vs. Reading Scores
plt.figure(figsize=(8, 6))
sns.scatterplot(x='math score', y='reading score', data=sd_data)
plt.title("Math vs. Reading Scores")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.show()
```

```python
# Visualization 4: Stacked Bar Chart - Parental Education Level by Test Preparation Course
plt.figure(figsize=(10, 6))
edu_prep = sd_data.groupby(['parental level of education', 'test preparation course']).size().unstack()
edu_prep.plot(kind='bar', stacked=True)
plt.title("Parental Education Level by Test Preparation Course")
plt.xlabel("Parental Education Level")
plt.ylabel("Count")
plt.show()
```

```python
# Visualization 5: Histogram - Writing Scores Distribution
plt.figure(figsize=(8, 6))
sns.histplot(sd_data['writing score'], bins=20, kde=True)
plt.title("Writing Scores Distribution")
plt.xlabel("Writing Score")
plt.ylabel("Frequency")
plt.show()
```
